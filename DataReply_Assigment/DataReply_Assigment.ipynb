```python
#!/usr/bin/python
file=open("Dergiler_Yorumlar.txt","r+",encoding="utf8")
my_wordcount_set={}
for word in file.read().split():
    if word not in my_wordcount_set:
        my_wordcount_set[word] = 1
    else:
        my_wordcount_set[word] += 1
for k,v in my_wordcount_set.items():
    print(k, v)
```
